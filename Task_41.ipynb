```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
import statsmodels.api as sm
```

```python
insurance = pd.read_csv('insurance.csv')
```

```python
insurance.columns
```

```python
#Plot a scatter plot with age on the x-axis and charges on the y-axis
plt.style.use('default')
sns.scatterplot(x = insurance['age'],y = insurance['charges'],)
plt.title('Relationship between age and charges')
plt.grid(False)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# check the dimensions of the fields in question
insurance['age'].shape
```

```python
insurance['charges'].shape
```

```python
# reshape the explanatory variable
X = insurance['age'].values.reshape(-1,1)
y = insurance['charges'].values
```

```python
# Using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on data.
linreg_model = LinearRegression()
linreg_model.fit(X, y)
y_pred = linreg_model.predict(X)
```

```python
# Plot another scatter plot with the best-fit line
sns.scatterplot(x = insurance['age'],y = insurance['charges'])
plt.plot(X,y_pred)
plt.title('Line of best fit beteeen age and charges')
plt.tight_layout()
plt.show()
plt.close()
```
